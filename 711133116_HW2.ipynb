```python
#第一題
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3*np.pi, 3*np.pi,200)
y = np.sin(x)+np.cos(x)
z=np.sin(x)
r=np.cos(x)
fig, ax = plt.subplots(1)
ax.plot(x, y,label='sin(x)+cos(x)',color='red')
ax.plot(x, z,label='cos(x)',color='y',linestyle = '--')
ax.plot(x, r,label='sin(x)',color='green',linestyle = 'dashdot')
ax.legend()
ax.set_xticks(np.array([-3, -2, -1, 0, 1, 2, 3])*np.pi)
ax.set_xticklabels(['-3$\pi$', '-2$\pi$', '$\pi$', '0', '$\pi$','2$\pi$','3$\pi$'], \
        fontsize=10, color = 'b')
ax.set_ylim([-3, 3])
ax.grid(True)
plt.show()
```

```python
#第二題
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 6, 100)
sig = (1 - np.exp((-2)*x)) / (1 + np.exp((-2)*x))

plt.axhline(y=1, color="b", linestyle="--")
plt.axhline(y=-1.0, color="purple", linestyle="--")
plt.plot(x, sig, color="gray", label=r"$f(x)=\frac{1-e^{-2 x}}{1+e^{-2 x}}$")
plt.xlim(-5, 6)
plt.grid(True)
plt.xlabel("x")
plt.legend(fontsize=14)
plt.show()
```

```python
#第三題
import matplotlib.pyplot as plt
import numpy as np
 
x = np.arange(-10, 10,0.100 ) 
# formulate a function f
f = lambda x : np.cbrt((4-x**3)/(1+x**2))
 
fig = plt.figure(figsize=[6, 4])
plt.plot(x, f(x), linestyle = 'dashdot', color = 'c')
plt.grid(visible = True, color='gray', linestyle='-', linewidth=0.5)
plt.xlabel('X'), plt.ylabel('Y')
plt.title(r'$f(x)=\sqrt[3]{\frac{4-x^{3}}{1+x^{2}}}$',size=15)
plt.show()
```

```python
#第五題
import matplotlib.pyplot as plt
import numpy as np
n = 100 # number of points
x = np.linspace(-20, 20, n) 
y = (1/(2*np.power(2*np.pi,1/2)))*np.power(np.e,-(((x-2)**2)/8))  
plt.plot(x, y)
plt.grid(True)
plt.xlabel('X')
plt.ylabel('Y')
def annot_max(x,y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = y.max()
    text= "x={:.3f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="bottom")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
annot_max(x,y) 
plt.title(r'$f(x)=\frac{1}{2 \sqrt{2 \pi}} e^{-\frac{(x-1)^{2}}{8}}$',size=20)
plt.show()  # just for py file
```

```python
#第七題
import matplotlib.pyplot as plt
import numpy as np
 
xmin, xmax = -2, 3
x = np.linspace(xmin, xmax, 100)
p = [-1, 2, 0, 0, 0]
f = lambda x : np.polyval(p, x)
 
fig = plt.figure()
# use axis object
ax = plt.gca() # get current axis
ax.plot(x, f(x), linewidth = 3, color ='g', linestyle ='-.')
ax.grid(True)
ax.set_xlabel('X'), ax.set_ylabel('f(x)')
ax.set_title('$f(x)=2x^3-x^4$')
# save the current figure as an eps file
plt.savefig('poly.eps', format='eps')
#反曲點
myfont = matplotlib.font_manager.FontProperties(fname=r'C:/Windows/Fonts/msjh.ttc')
plt.scatter([0, ], [0, ], s=50, color='b')
plt.text(0,-5,'(0,0)',fontdict={'size': 14, 'color': 'blue'},fontproperties=myfont)
plt.scatter([1, ], [1, ], s=50, color='b')
plt.text(1,-5,'(1,1) 反曲點',fontdict={'size': 14, 'color': 'blue'},fontproperties=myfont)
plt.show()
```

```python
#小專題-2
import numpy as np 
import matplotlib.pyplot as plt 
x = np.linspace(1, 10, 100)
f = lambda x : 1/x
#print(f(x))

plt.plot(x, f(x),color='brown',lw=2) 
ax = plt.gca()

s=[]
for i in x:
    a=1/int(i)
    #print(a)
    s.append(a)
ax.fill_between(x,s,f(x),where=s>=f(x),color='orange') 

# Re-arrange four axes spines
ax.spines['left'].set_position(('axes',0))
ax.spines['bottom'].set_position(('axes',0))
ax.spines['top'].set_visible(False) # take off top line
ax.spines['right'].set_visible(False)
ax.annotate(r'$f(x)=\frac{1}{x}$', xy=(1, 0.4))
ax.annotate(r'$\gamma_{n}=$ shaded  area', xy=(6, 0.4))

plt.show()
```

```python
from cProfile import label
import numpy as np 
import matplotlib.pyplot as plt 
x = np.linspace(1, 100, 100)
f = lambda x : 1+x*0
g = lambda x : (1/2)*(1-(1/(x+1)))

s=[]
for i in x:
    a=np.arange(1,i+1)
    Sn=(1/a).sum()
    s.append(Sn-np.log(i+1))
#print(s[0])
fig, ax = plt.subplots(figsize=(8,6))
# use axis object
#ax = plt.gca() # get current axis
ax.plot(x, f(x),label='1') 
ax.plot(x,s,label='$\gamma_n$')
ax.plot(x, g(x),label=r'$\frac{1}{2}\left(1-\frac{1}{n+1}\right)$')
plt.legend()
ax.grid(True)
ax.set_xlabel('X'), ax.set_ylabel('f(x)')

plt.show()
```

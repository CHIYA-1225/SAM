```python
Idx = (D_trian[:,2]==0) #boolean
plt.plot(D_trian[Idx, 0], D_trian[Idx, 1], "ro",\
alpha = 0.5, label = "Group A")

Idx = (D_trian[:,2]==1)
plt.plot(D_trian[Idx,0], D_trian[Idx,1],"bo", \
alpha = 0.5, label = "Group B")

plt.legend(), plt.grid(True)
plt.xlabel("$X_1$"), plt.ylabel("$X_2$")

# 3. Estimate the coefficients
n = len(D_trian[:, 0])
X = np.hstack((np.ones((n, 1)), D_trian[:, 0:2]))
y = D_trian[:, 2]
b = LA.inv(X.T @ X) @ X.T @ y.T
#b = LA.pinv(X) @ y
print(b)
# Draw a linear regression line
x = np.array([-2, 4.5])
y = -(b[0] - 0.5 + b[1] * x) / b[2]
plt.plot(x,y, lw=3)
plt.show()
```

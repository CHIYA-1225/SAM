```python
import matplotlib.pyplot as plt
import matplotlib
myfont = matplotlib.font_manager.FontProperties(fname=r'C:/Windows/Fonts/msjh.ttc')
```

```python
fig=plt.figure(figsize=[8,6])
plt.xlabel('',fontproperties=myfont,fontsize='20')
plt.ylabel('',fontproperties=myfont,fontsize='20')
LR=[0.8525,0.5625,0.7125,0.75]
AR=[0.8625,0.75,0.7625,0.8375]
LDA=[0.8625,0.575,0.7125,0.725]
QDA=[0.875,0.825,0.8375,0.9125]
KNN_15=[0.8875,0.825,0.825,0.8875]
KNN_5=[0.9,0.8125,0.875,0.9125]
ann_20=[0.775,0.7375,0.7625,0.7875]
ann_10=[0.785,0.7375,0.8125,0.8375]

x=['1','2','3','4']
plt.plot(x,LR,'--o',label='Linear Regression')
plt.plot(x,AR,'--o',label='Augmented Regression')
plt.plot(x,LDA,'--o',label='LDA')
plt.plot(x,QDA,'--o',label='QDA')
plt.plot(x,KNN_15,'--o',label='KNN_15')
plt.plot(x,KNN_5,'--o',label='KNN_5')
plt.plot(x,ann_20,'--o',label='ANN_20')
plt.plot(x,ann_10,'--o',label='ANN_10')
plt.legend(loc='best')
plt.grid()
plt.show()
```

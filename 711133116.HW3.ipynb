```python
#第一題-卡方
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
 
xlim = [0, 50]
x = np.linspace(xlim[0], xlim[1], 1000)
plt.figure()
plt.axis([xlim[0], xlim[1], 0, 0.2])

# df 
df = np.arange(4, 32, 2)

for i in df:
    y=chi2.pdf(x, i)
    plt.plot(x,y, lw=1, color='navy')

plt.plot(x,y, lw=1, color='navy')
plt.title(r'$\chi^2$ Distribution')
plt.yticks([0, 0.1, 0.2])
plt.show()
#自由度增加，圖形逐漸像右移動
```

```python
#第一題-t
import numpy as np
from scipy.stats import t
from scipy.stats import norm
import matplotlib.pyplot as plt
 
xlim = [-6,6]
x = np.linspace(xlim[0], xlim[1], 1000)
fig = plt.figure(figsize=[10,6])
plt.axis([xlim[0], xlim[1], 0, 0.41])

# df 
df1 = np.arange(3, 30, 3)
df2 = np.arange(0.1, 1, 0.1)
for i in df1:
    y=t.pdf(x, i)
    #print(i)
    plt.plot(x,y, lw=1, color='blue')
for i in df2:
    y=t.pdf(x, i)
    print(i)
    plt.plot(x,y, lw=1, color='blue')

plt.plot(x[::5],norm.pdf(x[::5]),label='Z distribution',color='red',lw=2)
plt.legend()
plt.title('t Distribution')

plt.show()
#隨自由度升高，漸近於標準常態
```

```python
#第一題-Beta
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2) 

xlim = [0,1]
x = np.linspace(xlim[0], xlim[1],1000)
 
ab= np.arange(1, 15)
for i in ab:
    y=beta.pdf(x,15,i)
    axs[0,0].plot(x,y, lw=1,color='blue') 
    y=beta.pdf(x,i,i)
    axs[1,0].plot(x,y, lw=1,color='red')
    y=beta.pdf(x,i,15)
    axs[0,1].plot(x,y, lw=1,color='green')  

axs[0,0].set_title(r'$a>b$')
axs[1,0].set_title(r'$a=b$')
axs[0,1].set_title(r'$a<b$')

ab2=np.arange(1,8,2)
ab3=np.arange(1,8,2)
for i in ab2:
    for z in ab3:
        y=beta.pdf(x,i,z)
        axs[1,1].plot(x,y, lw=1,color='orange')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs[0].flat:
    ax.label_outer()
plt.show()
```

```python
#第一題-負二項
import numpy as np
from scipy.stats import nbinom
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[12,6])
axs = fig.subplots(1, 3)

plt.suptitle("The PMF and CDF of Negative Binomial Distribution", fontsize="18")

X = np.arange(3, 30)
for i in range(5,20,5):
    r = i
    P = 0.5
    nbinom_pd = nbinom.pmf(X, r, P)
    axs[i//5-1].vlines(X, 0, nbinom_pd, colors='b', lw=5, alpha=0.5)
    nbinom_pd = nbinom.cdf(X, r, P)
    axs[i//5-1].plot(X,nbinom_pd,drawstyle = 'steps-pre')
    axs[i//5-1].set_title(r'n={}'.format(i),fontsize=15)
for ax in axs.flat:
    ax.label_outer()
fig.tight_layout()
```

```python
#第一題-負二項
import numpy as np
from scipy.stats import nbinom
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[12,6])
axs = fig.subplots(1, 3)

plt.suptitle("The PMF and CDF of Negative Binomial Distribution", fontsize="18")

X = np.arange(3, 30)
P = [0.1,0.5,0.8]
for i in range (3):
    r = 20
    nbinom_pd = nbinom.pmf(X, r, P[i])
    axs[i].vlines(X, 0,nbinom_pd, colors='g', lw=5, alpha=0.5)
    nbinom_pd = nbinom.cdf(X, r, P[i])
    axs[i].plot(X,nbinom_pd,drawstyle = 'steps-pre')
    axs[i].set_title(r'p={}'.format(P[i]),fontsize=15)

fig.tight_layout()
```

```python
#第一題-超幾何
import numpy as np
from scipy.stats import hypergeom
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[12,6])
axs = fig.subplots(1, 3)

for i in range(3):
    rv = hypergeom(40,10*(i+1),20)
    x = np.arange(0,20)
    pmf= rv.pmf(x)
    cdf= rv.cdf(x)
    axs[i].bar(x, pmf,color='brown')
    axs[i].plot(x, cdf,color='green')
    axs[i].set_title(r'n={}'.format(10*(i+1)),fontsize=15)
plt.suptitle("The PMF and CDF of Hypergeom Distribution", fontsize="18")
for ax in axs.flat:
    ax.label_outer()
fig.tight_layout()
```

```python
#第三題-常態機率圖 
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random

fig = plt.figure(figsize=[12,8])
axs = fig.subplots(2, 3)

x=[0,0,0,0,0,0]
x[0]=random.poisson(lam= 20 , size= 1000 ) 
x[1]=random.negative_binomial(1,0.5, size= 1000 )
x[2]=random.hypergeometric(40,20,20, size= 1000 )
x[3]=random.beta(3,9 , size= 1000 )
x[4]=random.gamma(3, size= 1000 )
x[5]=random.exponential(10, size= 1000 )

for i in range(6):
    if i<3:
        stats.probplot(x[i],plot=axs[0,i])
    else:
        stats.probplot(x[i],plot=axs[1,i-3])

axs[0,0].set_title('poisson',fontsize=16)
axs[0,1].set_title('negative_binomial',fontsize=16)
axs[0,2].set_title('hypergeometric',fontsize=16)
axs[1,0].set_title('beta',fontsize=16)
axs[1,1].set_title('gamma',fontsize=16)
axs[1,2].set_title('exponential',fontsize=16)

fig.tight_layout()
plt.suptitle('probability Plot',y=1.1,fontsize=20)
plt.show
```

```python
#第三題-經驗累積密度函數 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson 
from scipy.stats import nbinom
from scipy.stats import hypergeom
from scipy.stats import beta
from scipy.stats import gamma
from scipy.stats import expon
from scipy.stats import cumfreq  # for ECDF
import seaborn as sns
from numpy import random

fig = plt.figure(figsize=[12,8])
axs = fig.subplots(2, 3)

n = 1000
poi=random.poisson(lam= 20 , size= 1000 ) 
nb=random.negative_binomial(1,0.5, size= 1000 )
hg=random.hypergeometric(40,20,20, size= 1000 )
b=random.beta(3,9, size= 1000 )
g=random.gamma(3, size= 1000 )
exp=random.exponential(10, size= 1000 ) 

dis=[poi,nb,hg,b,g,exp]
num_bins = n 
for i in range(6):
    res = cumfreq(dis[i], num_bins)
    x = res.lowerlimit + np.linspace(0,res.binsize * res.cumcount.size, res.cumcount.size)
    if i<3:
        axs[0,i].plot(x, res.cumcount/n, drawstyle = 'steps-pre',label='Empirical CDF')
    else:
        axs[1,i-3].plot(x, res.cumcount/n, drawstyle = 'steps-pre')


x_ = np.sort(poi) 
poi_ = poisson.cdf(x_,20)
axs[0,0].plot(x_,poi_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')

x_ = np.sort(nb) 
nb_= nbinom.cdf(x_, 1, 0.5)
axs[0,1].plot(x_,nb_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')

x_ = np.sort(hg) 
hg_= hypergeom.cdf(x_,60,40,20)
axs[0,2].plot(x_,hg_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')

x_ = np.sort(b) 
beta_= beta.cdf(x_,3, 9)
axs[1,0].plot(x_,beta_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')

x_ = np.sort(g) 
gamma_= gamma.cdf (x_,3)
axs[1,1].plot(x_,gamma_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')

x_ = np.sort(exp)
Y = np.arange(1, n+1) / n 
#axs[1,2].plot(x, Y, drawstyle = 'steps-pre', label = 'Empirical CDF')
exp_= expon.cdf(x_,1,10)
axs[1,2].plot(x_,exp_, color = 'r', linestyle = '--',linewidth = 5, alpha = 0.5, label = 'real CDF')


axs[0,0].set_title('poisson',fontsize=16)
axs[0,1].set_title('negative_binomial',fontsize=16)
axs[0,2].set_title('hypergeometric',fontsize=16)
axs[1,0].set_title('beta',fontsize=16)
axs[1,1].set_title('gamma',fontsize=16)
axs[1,2].set_title('exponential',fontsize=16)

axs[0,0].grid(True)
axs[0,1].grid(True)
axs[0,2].grid(True)
axs[1,0].grid(True)
axs[1,1].grid(True)
axs[1,2].grid(True)
fig.tight_layout()
axs[0,0].legend()
plt.suptitle('The Empirical CDF with 1000 samples',y=1.1,fontsize=20)
plt.show
```
